## Como usar o ColorMap

VocÃª pode especificar o mapa de cores com o argumento de palavra-chave cmapcom o valor do mapa de cores, neste caso, 'viridis'que Ã© um dos mapas de cores integrados disponÃ­veis no Matplotlib.

# GrÃ¡ficos de pizza Matplotlib

```python
y = np.array([35, 25, 25, 15])

plt.pie(y)
plt.show()
```

Por padrÃ£o, a plotagem da primeira fatia comeÃ§a no eixo x e se move no sentido anti-horÃ¡rio.

## RÃ³tulos
Adicione rÃ³tulos ao grÃ¡fico de pizza com o labelparÃ¢metro.

O labelparÃ¢metro deve ser uma matriz com um rÃ³tulo para cada fatia:

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.show() 
```

## Explodir
Talvez vocÃª queira que uma das cunhas se destaque? O explodeparÃ¢metro permite que vocÃª faÃ§a isso.

O explodeparÃ¢metro, se especificado, e nÃ£o None, deve ser uma matriz com um valor para cada fatia.

Cada valor representa a que distÃ¢ncia do centro cada cunha Ã© exibida:

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]

plt.pie(y, labels = mylabels, explode = myexplode)
plt.show() 
```

## Legenda
Para adicionar uma lista de explicaÃ§Ã£o para cada cunha, use a legend()funÃ§Ã£o:

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.legend()
plt.show() 
```

```python
import numpy as np
import pandas as pd
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
dados = pd.read_csv('https://raw.githubusercontent.com/profandersonvanin01/python_introducao/main/housing.csv')
```

```python
dados = dados[['latitude','longitude']]
```

```python
dados.head()
```

```python
plt.figure(figsize=(15,10))
plt.scatter(dados['longitude'].values,
            dados['latitude'].values,
            s=5, c='blue',
            alpha = 0.5, zorder=5)
```

```python
def plot_map(lat1,lat2,lon1,lon2,xmap,ymap):
  plt.figure(figsize=(xmap,ymap))
  map = Basemap(projection='cyl', resolution='h',
                llcrnrlat=lat1, urcrnrlat=lat2,
                llcrnrlon=lon1, urcrnrlon=lon2)
  map.drawcoastlines()
  map.fillcontinents(color='palegoldenrod', lake_color='lightskyblue')
  map.drawmapboundary(fill_color='lightskyblue')
  map.drawparallels(np.arange(lat1, lat2 + 0.5, 2), labels=[1,0,0,0])
  map.drawmeridians(np.arange(lon1, lon2 + 0.5, 2), labels=[0,0,0,1])
  return map
```
